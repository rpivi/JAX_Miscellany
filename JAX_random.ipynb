```python
import jax
import jax.numpy as jnp
```

```python
#Ora vediamo la generazione di numeri casuali: è molto diversa in JAX
#si basa tutto sulla key, che è un numero intero che serve a generare numeri casuali in modo deterministico

key = jax.random.PRNGKey(0)
```

```python
#per generare da una gaussiana standard, usiamo jax.random.normal:
x = jax.random.normal(key)

print(x)

#generiamo un numero casuale uniforme tra 0 e 1:
x = jax.random.uniform(key)
print(x)

#il risultato è deterministico, se eseguiamo di nuovo il codice otterremo lo stesso risultato
#finchè non cambiamo la key, otterremo sempre lo stesso risultato
```

```python
key, subkey = jax.random.split(key) # key viene modificata, subkey è una nuova key generata a partire da key
print(key)
print(subkey)

#le due key sono diverse, quindi genereranno numeri casuali diversi
x1 = jax.random.normal(key)
x2 = jax.random.normal(subkey)
print(x1)
print(x2)

#se rieseguo il codice ottengo risultati diversi!!
#perchè ho cambiato la key, quindi i numeri casuali generati sono diversi
# cosa fa davvero split?
#split prende una key e ne genera due nuove, che sono indipendenti tra loro
# vecchia key  →  nuova key + subkey
```

```python
#array casuale:
key, subkey = jax.random.split(key)

x = jax.random.normal(subkey, shape=(5,))
print(x)
#ogni volta che eseguiamo il codice, otteniamo un array diverso, perchè la key è diversa
#shape specifica la forma dell'array, in questo caso un array di 5 elementi
#proviamo a generare un array di 3x3:
key, subkey = jax.random.split(key)
x = jax.random.normal(subkey, shape=(3, 3))
print(x)
```
