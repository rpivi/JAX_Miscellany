```python
import jax
import jax.numpy as jnp

# Verifica versione
print(jax.__version__)

# Verifica dispositivo disponibile
print(jax.devices())

# Test base
x = jnp.array([1.0, 2.0, 3.0])
print(x)
```

```python
#Unico modo per "modificare" un array
x = x.at[0].set(10)
print(x)
```

```python
#Operazioni Elementari
x = jnp.array([1.,2.,3.])
y = jnp.array([4.,5.,6.])

print(x+y)
print(x*y)
print(jnp.sum(x))
```

```python
x = jnp.array([1.,2.,3.])

print(x + 1)
```

```python
#le funzioni sono già vettorizzate
x = jnp.array([1.,2.,3.])

print(jnp.exp(x))
print(jnp.sin(x))
print(jnp.sqrt(x))
```

```python
#egualmente se la funzione è definita da noi, è già vettorizzata
def f(x):
    return x**2 + 3*x

print(f(2.))
print(f(jnp.array([1.,2.,3.])))
```

```python
#Ora vediamo l'utilià di JAX nelle derivate automatiche
from jax import grad
```

```python
def V(x):
    return x**4 - 2*x**2
```

```python
dV = grad(V)

print(dV(1.0))
```

```python
#con jax jit non si può usare if quindi si usa where:
def f(x):
    return jnp.where(x > 0, x**2, -x**2)
print(f(2.0))
print(f(-2.0))

# where è nella forma(condizione, valore se vero, valore se falso)
```
