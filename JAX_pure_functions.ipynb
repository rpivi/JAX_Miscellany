```python
#una funzione è pura se dato lo stesso input da sempre lo stesso output e non ha effetti collaterali

#esempio di funzione pura
def somma(a, b):
    return a + b

#esempio di funzione impura
a = 0

def g(x):
    global a
    a = a + 1
    return x**2
#la funzione non è pura perchè modifica la variabile globale a ogni chiamata, quindi restituisce sempre lo stesso output per lo stesso input
#ma ha effetti collaterali sulla variabile globale a
#lo stesso input non produce lo stesso stato globale

#il random non è una funzione pura perchè restituisce un output diverso ogni volta che viene chiamata, anche se l'input è lo stesso
```

```python
#per superare il problema delle funzioni impure bisogna igegnarsi

import jax
import jax.numpy as jnp

def generate_random(key):
    key, subkey = jax.random.split(key)
    x = jax.random.normal(subkey)
    return x, key

#vediamo che è pura:
key = jax.random.PRNGKey(0)
x, key = generate_random(key)
print(x)
key = jax.random.PRNGKey(0)
x, key = generate_random(key)
print(x)

#stessa key stessa output, ma non ha effetti collaterali perchè la key viene restituita e non modificata globalmente
#questa funzione utilizza una chiave, la "consuma" e  restituisce il risultato e la nuova chiave, in questo modo è pura e non ha effetti collaterali

```

```python
#vediamo un loop che usa la funzione pura def precedentemente:
key = jax.random.PRNGKey(0)

for i in range(5):
    x, key = generate_random(key)
    print(x)

#schema:
#key0 → x1 + key1
#key1 → x2 + key2
#key2 → x3 + key3
```

```python
#una cofigurazione D dimensionale casuale:
def generate_config(key, D):
    key, subkey = jax.random.split(key)
    x = jax.random.normal(subkey, shape=(D,))
    return x, key
```

```python
#usiamola
key = jax.random.PRNGKey(0)
D = 3
config, key = generate_config(key, D)
print(config)
```
